### Sample Size vs. Median

```python
ax = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax.set_title('Sample size vs. Median')
plt.show()
```

### Sample Size vs. Unemployment Rate

```python
ax = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax.set_title('Sample Size vs. Unemployment Rate')
plt.show()
```

### Full Time vs. Median

```python
ax = recent_grads.plot(x='Full_time', y='Sample_size', kind='scatter')
ax.set_title('Full Time vs. Sample Size')
plt.show()
```

### Share Women vs. Unemployment Rate

```python
ax = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax.set_title('Share Women vs. Unemployment Rate')
plt.show()
```

### Men vs. Median

```python
ax = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax.set_title('Men vs. Median')
plt.show()
```

### Women vs. Median

```python
ax = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax.set_title('Women vs. Median')
plt.show()
```

### Do Students in more popular majors make more money?

```python
ax = recent_grads.plot(x='Total', y="Median", kind='scatter')
ax.set_title('Total number of people with Major vs. Median Salary')
plt.show()
```

The scatter plot doesn't indicate any strong patterns between Median salary and the Total number of people with that Major (an indication of its popularity) to suggest whether popular majors make more money.

### Do students that majored in subjects that were majority female make more money?

```python
ax = recent_grads.plot(x='ShareWomen', y="Median", kind='scatter')
ax.set_title('Women as Share of Total vs. Median Salary')
plt.show()
```

The scatter plot seems to indicate a slight negative correlation to Share of Women of Total and Median Salary, which indicates that the more popular the major is with women the lower the median salary. 

### Is there any link between the number of full-time employees and median salary?

```python
ax = recent_grads.plot(x='Full_time', y="Median", kind='scatter')
ax.set_title('Number employed 35 hours or more vs Median salary')
plt.show()
```

There doesn't seem to be any strong correlation between the number employed 35 hours and more with an increase in median salary.
